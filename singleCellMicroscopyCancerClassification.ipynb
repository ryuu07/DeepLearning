```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import torch.optim as optim
```

```python
df = pd.DataFrame(ds['train'])
print(df.info())
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['image'], df['label'], test_size=0.2, random_state=42)
```

```python
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import numpy as np
```

```python
class preProcessDataTrain(Dataset):
    def __init__(self,x_train,y_train):
        self.xTrain = []
        for img in x_train:
            if img.mode != 'RGB':
                img = img.convert('RGB')
            self.xTrain.append(img)
        self.yTrain = y_train

        self.preprocess = transforms.Compose([
            transforms.Resize((256,256)),
            transforms.RandomHorizontalFlip(),
            transforms.RandomVerticalFlip(p=0.5),       # Added: Essential for microscopy
            transforms.RandomRotation(degrees=180),
            transforms.ToTensor(),
            # transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    
    def __len__(self):
        return len(self.xTrain)
    
    def __getitem__(self, index):
        image,label = self.xTrain[index],self.yTrain[index]
        return self.preprocess(image), torch.tensor(label)
```

```python
class preProcessDataTest(Dataset):
    def  __init__(self,x_test,y_test):
        self.xTest= []
        for img in x_test:
            if img.mode != 'RGB':
                img = img.convert('RGB')
            self.xTest.append(img)
        self.yTest = y_test

    
        self.preprocess = transforms.Compose([
                transforms.Resize((256,256)),
                transforms.ToTensor(),
                # transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    
    def __len__(self):
        return len(self.xTest)
    
    def __getitem__(self, index):
        image,label = self.xTest[index],self.yTest[index]
        return self.preprocess(image), torch.tensor(label)
```

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        #stride 1, padding 1, padding_mode zeros, kernel_size 3, no of filters 8, image channels 3
        self.conv1 = nn.Conv2d(stride=1, padding=1, padding_mode='zeros',kernel_size=3, out_channels=32, in_channels=3)
        # self.bn1 = nn.BatchNorm2d(32) # why the same number of filters as conv layer? because each filter  represents a specific feature like an edge.
        self.maxpool1 = nn.MaxPool2d(stride=2, kernel_size=2)
        self.conv2 = nn.Conv2d(stride=1, padding=1, padding_mode='zeros',kernel_size=3, out_channels=64, in_channels=32)
        # self.bn2 = nn.BatchNorm2d(64)
        self.maxpool2 = nn.MaxPool2d(stride=2, kernel_size=2)
        self.conv3 = nn.Conv2d(stride=1, padding=1, padding_mode='zeros',kernel_size=3, out_channels=128, in_channels=64)
        # self.bn3 = nn.BatchNorm2d(128)
        self.maxpool3 = nn.MaxPool2d(stride=2, kernel_size=2)
        self.conv4 = nn.Conv2d(stride=1, padding=1, padding_mode='zeros',kernel_size=3, out_channels=256, in_channels=128)
        # self.bn4 = nn.BatchNorm2d(256)
        self.maxpool4 = nn.MaxPool2d(stride=2, kernel_size=2)
        self.flatten = nn.Flatten()
        self.fully_connected_network = nn.Sequential(
            nn.Linear(in_features=256*16*16, out_features=512),
            nn.ReLU(),
            nn.Dropout(0.5), #<---  added dropout  to reduce overfitting coz large number of neurons i.e. 65536 in the flatten layer

            nn.Linear(in_features=512, out_features=256),
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(in_features=256, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=1)
        )

    def forward(self, x):
        #layer1
        z1 = self.conv1(x)
        # b1 = self.bn1(z1) #batchNormalization
        A1 = F.relu(z1)
        P1 = self.maxpool1(A1)

        #layer2
        z2 = self.conv2(P1)
        # b2 = self.bn2(z2)
        A2 = F.relu(z2)
        P2 = self.maxpool2(A2)

        #layer3
        z3 = self.conv3(P2)
        # b3 = self.bn3(z3)
        A3 = F.relu(z3)
        P3 = self.maxpool3(A3)

        #layer4
        z4 = self.conv4(P3)
        # b4 = self.bn4(z4)
        A4 = F.relu(z4)
        P4 = self.maxpool4(A4)

        #flatten
        outputs = self.flatten(P4) #256*16*16
        
        #fully-connected layer
        y_hat = self.fully_connected_network(outputs)

        return y_hat

```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CNN().to(device)
loss_fn = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(lr=0.0001,params=model.parameters())
```

```python
x_train_np = np.array(x_train)
y_train_np = np.array(y_train)
x_test_np = np.array(x_test)
y_test_np = np.array(y_test)
```

```python
train_dataset = preProcessDataTrain(x_train_np, y_train_np)
test_dataset = preProcessDataTest(x_test_np,y_test_np)
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=32,shuffle=True)
```

```python
epochs = 10
for epoch in range(epochs):
    model.train()
    total_loss_per_epoch = 0
    
    # Standardizing naming to 'batch_labels'
    for batch_features, batch_labels in train_loader:

        # Move both to GPU
        batch_features = batch_features.to(device)
        batch_labels = batch_labels.to(device)
        
        # 1. Forward pass
        y_hat = model(batch_features)
        
        # 2. Match dimensions: y_hat is [batch, 1], labels must be [batch, 1]
        # and both must be float
        loss = loss_fn(y_hat, batch_labels.unsqueeze(1).float())

        # 3. Backward pass
        optimizer.zero_grad()
        loss.backward()

        # 4. Update weights 
        optimizer.step()
        
        total_loss_per_epoch += loss.item()
    
    avg_loss = total_loss_per_epoch / len(train_loader)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")
```

```python
model.eval()
total = 0
correct = 0 

with torch.no_grad():
    for batch_features, batch_labels in test_loader:
        batch_features = batch_features.to(device)
        batch_labels = batch_labels.to(device)
        
        # 1. Get raw logits from model
        outputs = model(batch_features).squeeze() # Turn [batch, 1] into [batch]
        
        # 2. If logit > 0, it's Class 1. If logit <= 0, it's Class 0.
        predicted = (outputs > 0.0).float()
        
        # 3. Count accuracy
        total += batch_labels.size(0)
        correct += (predicted == batch_labels).sum().item()

print(f"Real Test accuracy: {100 * correct / total:.2f}%")
```
